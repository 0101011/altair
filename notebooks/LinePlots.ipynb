# Line Plots

```python
from altair import *
```

```python
data = load_dataset('stocks')
```

## Simple Line Chart

### Data Transform in Vega-Lite

```python
Chart(data).mark_line().encode(
    x='date:T',
    y='price:Q'
).transform_data(filter="datum.symbol==='GOOG'")
```

### Data Transform in Pandas

```python
Chart(data.query('symbol == "GOOG"')).mark_line().encode(
    x='date:T',
    y='price:Q'
)
```

## Multiple-Line Chart

```python
Chart(data).mark_line().encode(
    x='date:T',
    y='price',
    color='symbol',
)
```

```python
Chart(data).mark_line().encode(
    x='date:T',
    y='price',
    detail='symbol',
)
```

## Line Chart with Custom Path

```python
data = load_dataset('driving')
Chart(data).mark_line().encode(
    X('miles', scale=Scale(zero=False)),
    Y('gas', scale=Scale(zero=False)),
    path='year:T',
)
```
