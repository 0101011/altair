# Line Plots

```python
from altair import *
```

## Simple Line Chart

### Data Transform in Vega-Lite

```python
data = load_dataset('stocks')

# Transform in Vega-Lite
layer = Layer(data, transform=Transform(filter="datum.symbol==='GOOG'"))
layer.encode(x='date:T',
             y='price:Q')
layer.mark_line()
```

### Data Transform in Pandas

```python
# Transform in Pandas
layer = Layer(data.query('symbol == "GOOG"'))
layer.encode(x='date:T',
             y='price:Q')
layer.mark_line()
```

## Multiple-Line Chart

```python
data = load_dataset('stocks')
Layer(data).encode(
    x='date:T',
    y='price',
    color='symbol',
).mark_line()
```

```python
data = load_dataset('stocks')
Layer(data).encode(
    x='date:T',
    y='price',
    detail='symbol',
).mark_line()
```

## Line Chart with Custom Path

```python
data = load_dataset('driving')
Layer(data).encode(
    X('miles', scale=Scale(zero=False)),
    Y('gas', scale=Scale(zero=False)),
    path='year:T',
).mark_line()
```
