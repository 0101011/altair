# Scatter-Plot Examples

```python
from altair import *

cars = load_dataset('cars')
```

## Dot Plot

```python
Layer(cars).encode(
    x='Horsepower',
).mark_point()
```

## Scatter Plot

```python
Layer(cars).encode(
    x='Horsepower',
    y='Miles_per_Gallon'
).mark_point()
```

## Bubble Plot

```python
Layer(cars).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    size='Acceleration'
).mark_point()
```

## Scatter Plot with Color and/or Shape

```python
Layer(cars).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    shape='Origin',
).mark_point()
```

## Circle Scatterplot

```python
Layer(cars).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
).mark_circle()
```

## Square ScatterPlot

```python
Layer(cars).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
).mark_square()
```

## Text ScatterPlot

```python
Layer(cars).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    text='Origin'
).mark_text()
```

## Tick Plot

```python
Layer(cars).encode(
    x='Horsepower:Q',
    y='Cylinders:O',
).mark_tick()
```
