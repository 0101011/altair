# Area Plots

```python
from altair import *
```

```python
data = load_dataset('unemployment-across-industries')
data['date'] = data['date'].astype(str)
```

## Simple Area Chart

```python
Layer(data).encode(
    X('date:T', axis=Axis(format='%Y', labelAngle=0)),
    y='sum(count)'
).mark_area()
```

## Stacked Area Chart

```python
Layer(data).encode(
    X('date:T',
      timeUnit='yearmonth',
      axis=Axis(format='%Y', labelAngle=0, title='year')),
    Y('sum(count)'),
    Color('series:N', scale=Scale(range="category20b"))
).mark_area()
```

## Normalized Area Chart

```python
Layer(data).encode(
    X('date:T', timeUnit='yearmonth',
      axis=Axis(format='%Y', labelAngle=0, title='year')),
    Y('sum(count)'),
    color=Color('series:N', scale=Scale(range="category20b"))
).configure(
    mark=MarkConfig(stacked='normalize')  
).mark_area()
```

## SteamGraph

```python
Layer(data).encode(
    X('date:T', timeUnit='yearmonth',
      axis=Axis(format='%Y', labelAngle=0, title='year')),
    Y('sum(count)'),
    color=Color('series:N', scale=Scale(range="category20b"))
).configure(
    mark=MarkConfig(stacked='center')  
).mark_area()
```
